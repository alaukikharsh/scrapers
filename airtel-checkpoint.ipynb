```python
def get_stores_by_pincode(driver, pincode):
    driver.find_element_by_id('autocomplete').clear()
    driver.find_element_by_id('autocomplete').send_keys(pincode)
    driver.find_element_by_id('autocomplete').send_keys(u'\ue007')
    sleep(2)
    pg=bs(driver.page_source)
    pointers=driver.find_elements_by_css_selector("a[ng-click=\"vm.toggleMarker(\'Airtel Store\');\"]")
    class_name=pointers[0].get_attribute("class")
    if class_name=="btn-circle btn-sm tags":
        driver.find_elements_by_css_selector("a[ng-click=\"vm.toggleMarker(\'Airtel Store\');\"]")[0].click()
    pointers=driver.find_elements_by_css_selector("a[ng-click=\"vm.toggleMarker(\'Payment Bank\');\"]")
    class_name=pointers[0].get_attribute("class")
    if class_name=="btn-circle btn-sm tags selected":
        driver.find_elements_by_css_selector("a[ng-click=\"vm.toggleMarker(\'Payment Bank\');\"]")[0].click()
    pointers=driver.find_elements_by_css_selector("a[ng-click=\"vm.toggleMarker(\'Airtel Express\');\"]")
    class_name=pointers[0].get_attribute("class")
    if class_name=="btn-circle btn-sm tags selected":
        driver.find_elements_by_css_selector("a[ng-click=\"vm.toggleMarker(\'Airtel Express\');\"]")[0].click()
```

```python
def get_store_addresses(driver,source, airtel_stores,latitude,longitude):
    page = bs(source)
    addresses = page.find_all('div', {'class': 'fs-14'})
    j=0
    for i in range(len(addresses)):
        address = addresses[i].text
        if address in airtel_stores:
            continue
        airtel_stores.append(address)
        pointers=driver.find_elements_by_id("storeD"+str(j))
        pointers[0].click()
        #poi=driver.find_elements_by_title("Open this area in Google Maps (opens a new window)")
        poi=driver.find_element_by_xpath('//*[@title="Open this area in Google Maps (opens a new window)"]')
        ll=poi.get_attribute('href')[32:-40]
        lat=ll.split(',',2)[0]
        long=ll.split(',',2)[1]
        latt=open('C:/Users/abdenather/Desktop/xelpmoc/airtel stores/lat.txt','a+',encoding="utf-8")
        latt.write(lat+"\n")
        latt.close()
        longg=open('C:/Users/abdenather/Desktop/xelpmoc/airtel stores/long.txt','a+',encoding="utf-8")
        longg.write(long+"\n")
        longg.close()
        stores=open('C:/Users/abdenather/Desktop/xelpmoc/airtel stores/address.txt','a+',encoding="utf-8")
        stores.write(address+"\n")
        stores.close()
        #break
        j+=1
```

```python
airtel_stores = []
latitude=[]
longitude=[]
jj=791
for i in pincodes[791:]:
    get_stores_by_pincode(driver, i)
    sleep(3)
    get_store_addresses(driver,driver.page_source, airtel_stores, latitude,longitude)
    sleep(3)
    print(jj)
    jj+=1
```
