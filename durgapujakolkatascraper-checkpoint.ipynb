```python
import csv
import json
import re
import selenium
import requests
from selenium import webdriver
import pandas as pd
import numpy as np
from time import sleep
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
```

```python
driver = selenium.webdriver.Chrome(executable_path='C:/Users/abdenather/Desktop/chromedriver_win32/chromedriver.exe')
```

```python
driver.get('https://wbtourismpuja.in/home/puja_list')
```

```python
Name=[]
Organiser=[]
Address=[]
Description=[]
lat=[]
long=[]
content_element = driver.find_element_by_class_name("pujaListView")
content_html = content_element.get_attribute("innerHTML")
soup = bs(content_html, "html.parser")
p_tags = soup.find_all("a")
j=0
for i in p_tags:
    url1=requests.get(i['href'])
    source1 = bs(url1.text, 'html.parser')
    span5=source1.findAll("script")[7].text
    lat_ind=span5.find('lat')
    lng_ind=span5.find('lng')
    span6=span5[lat_ind+7:lat_ind+16]   
    span7=span5[lng_ind+7:lng_ind+16]
    try:
        latt=(float(span6))
        longg=(float(span7))
    except:
        try:
            latt=(float(span6[:7]))
            longg=(float(span7[:7]))
        except:
            w=1
            print(j)
            j+=1
            continue
    lat.append(latt)
    long.append(longg)
    span=source1.findAll("span", {"id": "ContentPlaceHolder1_lblPujaName"})[0].text
    span2=source1.findAll("span", {"id": "ContentPlaceHolder1_lblOrganiser"})[0].text
    span3=source1.findAll("span", {"id": "ContentPlaceHolder1_lblAddress"})[0].text
    span4=source1.findAll("span", {"id": "ContentPlaceHolder1_lblDescription"})[0].text
    Name.append(span)
    Organiser.append(span2)
    Address.append(span3)
    Description.append(span4)
    print(j)
    j+=1
```
