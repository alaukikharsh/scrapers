```python
ii=0
read_more=driver.find_element_by_xpath('/html/body/div[3]/section[1]/div[3]/h3')
while ii<30:
    driver.execute_script("arguments[0].scrollIntoView();", read_more)
    sleep(5)
    ii+=1
```

```python
a=driver.find_elements_by_tag_name('a')
j=0
urls=[]
k=1
for i in a[24:-132]:
    if j%6==0:
        try:
            elem=driver.find_element_by_xpath('//*[@id="profile-listing"]/div[2]/ul/li[{0}]/div/div[2]/div[1]/div[2]/ul'.format(k)).text
        except:
            k+=1
            elem=driver.find_element_by_xpath('//*[@id="profile-listing"]/div[2]/ul/li[{0}]/div/div[2]/div[1]/div[2]/ul'.format(k)).text
        prop=elem.splitlines()
        if len(prop)>=7:
            name=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/Name.txt','a+',encoding="utf-8")
            name.write(i.text[:i.text.find("LV-")]+"\n")
            name.close()
            agenheight=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/agenheight.txt','a+',encoding="utf-8")
            agenheight.write(prop[0]+"\n")
            agenheight.close()
            Religion=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/religion.txt','a+',encoding="utf-8")
            Religion.write(prop[1]+"\n")
            Religion.close()
            Study=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/Study.txt','a+',encoding="utf-8")
            Study.write(prop[2]+"\n")
            Study.close()
            Marital=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/Marital.txt','a+',encoding="utf-8")
            Marital.write(prop[3]+"\n")
            Marital.close()
            Address=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/Address.txt','a+',encoding="utf-8")
            Address.write(prop[4]+"\n")
            Address.close()
            Proffession=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/Proffession.txt','a+',encoding="utf-8")
            Proffession.write(prop[5]+"\n")
            Proffession.close()
            Language=open('C:/Users/abdenather/Desktop/xelpmoc/matrimonial/Language.txt','a+',encoding="utf-8")
            Language.write(prop[6]+"\n")
            Language.close()
            print(i.text[:i.text.find("LV-")],prop[6])
        k+=1
    j+=1
```
