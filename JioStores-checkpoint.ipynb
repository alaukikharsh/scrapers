```python
def get_stores_by_pincode(driver, pinCode):
    driver.find_element_by_class_name('locationInputText').clear()
    driver.find_element_by_class_name('locationInputText').send_keys(pinCode)
    driver.find_element_by_class_name('locationInputText').send_keys(u'\ue007')
```

```python
def get_all_stores(driver):
    action = ActionChains(driver)    
    action.move_to_element(driver.find_element_by_class_name('showAllAppsButton'))
    sleep(1)
    while True:
        try:
            driver.find_element_by_class_name('showAllAppsButton').click()
            sleep(1)
        except Exception as e:
            print(e)
            break
```

```python
def get_store_addresses(source, jio_stores, unique_stores):
    page = bs(source)
    addresses = page.find_all('span', {'class': 'stroeaddress'})
    rejected = []
    for i in range(len(addresses)):
        address = addresses[i]['title']
        if address in unique_stores:
            rejected.append(i)
            continue
        unique_stores.append(address)
        jio_stores.append({'address': address})
        stores=open('C:/Users/abdenather/Desktop/xelpmoc/jio stores/address.txt','a+',encoding="utf-8")
        stores.write(address+"\n")
        stores.close()
    return rejected
```

```python
def point(pointers,i):
    try:
        pointers[i].click()
    except:
        sleep(1)
        print("trying")
        driver.execute_script("window.scrollTo(0, 0)") 
        sleep(1)
        point(pointers,i)
```

```python
def get_store_locations(source, jio_stores, rejected):
    pointers = driver.find_elements_by_class_name('storeLocatorPointer')
    action = ActionChains(driver).move_to_element(driver.find_element_by_tag_name('body'))
    #action.move_to_element(driver.find_element_by_tag_name('body'))
    i=0
    j=0
    while i<len(pointers):#in range(len(pointers)):
        if i in rejected:
            i+=1
            continue
        i+=1
        action = ActionChains(driver).move_to_element(pointers[i-1])
        sleep(3)
        #pointers[i].click()
        point(pointers,i-1)
        page = bs(driver.page_source)
        lat_long_string = re.findall(r'\d\d.\d+_\d\d.\d+',page.find('a', {'id': 'go'})['href'])[0]
        lat, long = lat_long_string.split('_')
        jio_stores[j]['lat'] = lat
        stores_lat=open('C:/Users/abdenather/Desktop/xelpmoc/jio stores/address_lat.txt','a+',encoding="utf-8")
        stores_lat.write(lat+"\n")
        stores_lat.close()
        jio_stores[j]['long'] = long
        stores_long=open('C:/Users/abdenather/Desktop/xelpmoc/jio stores/address_long.txt','a+',encoding="utf-8")
        stores_long.write(long+"\n")
        stores_long.close()
        sleep(1)
        driver.find_element_by_class_name('gm-ui-hover-effect').click()
        sleep(1)
        pointers = driver.find_elements_by_class_name('storeLocatorPointer')
        j+=1
```

```python
jio_stores = []
unique_stores = []
#ppp=[475]
ind=795
for i in pincodes[795:]:
    get_stores_by_pincode(driver, i)
    sleep(1)
    get_all_stores(driver)
    jio_stores_per_pincode = []
    sleep(1)
    rejected = get_store_addresses(driver.page_source, jio_stores_per_pincode, unique_stores)
    sleep(1)
    get_store_locations(driver, jio_stores_per_pincode, rejected)
    print(ind)
    ind+=1
    jio_stores += jio_stores_per_pincode
```
