```python
stores_address=[]
contact=[]
latitude=[]
longitude=[]
url='https://www.ifbappliances.com/storelocator'
driver.get(url)
```

```python
select = Select(driver.find_element_by_id('storetype'))
select.select_by_value('2')
sleep(1)
select = Select(driver.find_element_by_id('state'))
select.select_by_value('Maharashtra')
sleep(1)
```

```python
element=driver.find_element_by_id("city")
cities=element.text
cities=[s.strip() for s in cities.splitlines()]
cities=cities[1:]
```

```python
cities
```

```python
i=0
driver.get(url)
select = Select(driver.find_element_by_id('storetype'))
select.select_by_value('2')
sleep(1)
select = Select(driver.find_element_by_id('state'))
select.select_by_value('Maharashtra')
sleep(1)
urls=[]
while i<len(cities):  
    select = Select(driver.find_element_by_id('city'))
    select.select_by_value(cities[i])
    print(cities[i],i)
    sleep(3)
    b=driver.find_element_by_id("store-search")
    #print(b)
    b.click()
    #break
    sleep(3)
    items=driver.find_elements_by_class_name('store-link-btn')
    for j in items:
        a=j.get_attribute('href')
        urls.append(a)
        print(a)
    sleep(2)
    i+=1
    #print("j")
```

```python
i=0
latitude=[]
longitude=[]
address=[]
contact=[]
name=[]
while i<len(urls):
    url=urls[i]
    url=url[:30]+"storelocator/"+url[30:]
    driver.get(url)
    elem=driver.find_element_by_id('latitude')
    latitude.append(elem.get_attribute('value'))
    print(elem.get_attribute('value'))
    elem=driver.find_element_by_id('longitude')
    longitude.append(elem.get_attribute('value'))
    print(elem.get_attribute('value'))
    elem=driver.find_element_by_class_name('each-store-address')
    address.append(elem.text)
    print(elem.text)
    elem=driver.find_element_by_class_name('each-store-name')
    name.append(elem.text)
    print(elem.text)
    elem=driver.find_element_by_class_name('wd50').text
    print(elem)
    contact.append(elem)
    print("dd")
    i+=1
```
